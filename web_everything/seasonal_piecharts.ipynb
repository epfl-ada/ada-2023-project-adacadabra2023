```python
filtered_df = df[['category','Season', 'country_user']]
filtered_df
```

```python
top_4_beerstyles = filtered_df.groupby(['country_user', 'Season'])['category'].value_counts().groupby(['country_user', 'Season']).head(4).reset_index(name='counts')
```

```python
#top_4_beerstyles['categories'] = top_4_beerstyles['style'].str.strip().map(beer_dict)
top_4_beerstyles
```

```python
summer_top4 = top_4_beerstyles[top_4_beerstyles['Season'] == 'Summer']
winter_top4 = top_4_beerstyles[top_4_beerstyles['Season'] == 'Winter']
autumn_top4 = top_4_beerstyles[top_4_beerstyles['Season'] == 'Autumn']
spring_top4 = top_4_beerstyles[top_4_beerstyles['Season'] == 'Spring']
summer_top4
```

```python
summer_reviews = pd.crosstab(summer_top4['country_user'], summer_top4['category'], 
                             values = summer_top4['counts'], aggfunc= 'mean').fillna(0)
winter_reviews = pd.crosstab(winter_top4['country_user'], winter_top4['category'], 
                             values = winter_top4['counts'], aggfunc= 'mean').fillna(0)
autumn_reviews = pd.crosstab(autumn_top4['country_user'], autumn_top4['category'], 
                             values = autumn_top4['counts'], aggfunc= 'mean').fillna(0)
spring_reviews = pd.crosstab(spring_top4['country_user'], spring_top4['category'], 
                             values = spring_top4['counts'], aggfunc= 'mean').fillna(0)
spring_reviews
```

```python
#Interactive plot
def seasonal_piechart (season_df):
    app = Dash(__name__)


    app.layout = html.Div([
        html.H4('Top4 beers ratings distribution in summer'),
        dcc.Graph(id="graph"),
        html.P("Countries:"),
        dcc.Dropdown(id='names',
            options=season_df.T.columns.values,
            value='Switzerland', clearable=False
        ),
    ])


    @app.callback(
        Output("graph", "figure"), 
        Input("names", "value"))
    def generate_chart(names):
        df = season_df.T # replace with your own data source
        fig = px.pie(df, values= names , names= season_df.T.index.values , hole=.3)
        fig.update_traces(textposition = 'inside')
        return fig


    app.run_server(debug=True)
```
